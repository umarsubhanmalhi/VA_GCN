```python
cat_name="Women"
subcat_name="Women_also_viewed"
cat_rel="also_viewed"
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Define the category and subcategory names (assuming these are consistent across files)
# List of k values and corresponding colors and line styles for better differentiation
k_values = [1, 5, 10, 20]
colors = ['red', 'orange', 'green', 'blue']  # More vivid colors
# List of k values and corresponding colors and line styles for better differentiation
line_styles = ['-', '--', '-.', ':']  # Distinct line styles

# Prepare the plot
plt.figure(figsize=(10, 8))  # Adjust size for better visibility
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Chance (area = 0.50)')  # Diagonal line

# Loop over each k value
for k, color, ls in zip(k_values, colors, line_styles):
    # Load prediction and label data
    pred_path = f'D:/School/FARS/Dataset/{cat_name}/{subcat_name}/pred_{k}.npy'
    label_path = f'D:/School/FARS/Dataset/{cat_name}/{subcat_name}/test_labels_{k}.npy'
    pred = np.load(pred_path)
    test_labels = np.load(label_path)

    # Compute ROC curve and ROC area
    fpr, tpr, _ = roc_curve(test_labels, pred)
    roc_auc = auc(fpr, tpr)

    # Plot the ROC curve
    plt.plot(fpr, tpr, color=color, lw=2, linestyle=ls, label=f'ROC curve for k={k} (area = {roc_auc:.2f})')

# Customize plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)
plt.title(f'ROC Curves for {subcat_name}', fontsize=24)
plt.legend(loc="lower right", fontsize=18)
plt.grid(True)
plt.show()
```
